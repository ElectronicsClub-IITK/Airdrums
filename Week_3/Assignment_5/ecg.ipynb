```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks
```

importing necessary lib

```python
file_path = r"C:\Users\rohna\Music\ecg\sam_ECG.csv" #reading and then loting data
data = pd.read_csv(file_path)
def plot(data):
    plt.plot(data['time'],data['ecg'])
    plt.xlabel("time")
    plt.ylabel("ecg")
    plt.xlim(800,900)
    plt.show()
plot(data)
```

ploting the data

```python
volt=data['ecg']
time=data['time']
```

storing in variables

```python
peaks, _= find_peaks(volt,prominence=1)#in g,f finding peaks
plt.subplot(2, 1, 2)
plt.plot(time,volt)
plt.scatter(time[peaks], volt[peaks], color='red', label='Peaks')
plt.title('ecg')
plt.xlabel('time')
plt.ylabel('voltage')
plt.show()
```
